```python
!pip install scikit-fuzzy
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

# Define the range for the universe of discourse
x = np.arange(0, 100, 10)

# Define fuzzy set "Slow" using a trapezoidal membership function
Slow = fuzz.trapmf(x, [0, 10, 20, 30])

# Define fuzzy set "Medium" using a trapezoidal membership function
Medium = fuzz.trapmf(x, [30, 40,50, 60])

# Define fuzzy set "Fast" using a trapezoidal membership function
Fast = fuzz.trapmf(x, [60, 70, 80, 90])

# Plot fuzzy sets "Slow", "Medium", and "Fast"
plt.figure(figsize=(40, 10))
plt.plot(x, Slow, label='Slow', linewidth=2)
plt.plot(x, Medium, label='Medium', linewidth=2)
plt.plot(x, Fast, label='Fast', linewidth=2)
plt.xlabel('Speed (Kmph)')
plt.ylabel('Membership Degree')
plt.title('Fuzzy Sets for Speed Levels')
plt.legend()
plt.grid(True)
plt.show()
```
