```python
import numpy as np
import matplotlib.pyplot as plt

#Define the range for the universe of discourse
x=np.arange(0,40,0.1)

#Define fuzzy set "cold" using a trpezoidal membership function
cold=np.maximum(0,np.minimum(1,np.minimum((15-x)/(15-10),1)))

#Define fuzzy set "Warm" using a trpezoidal membership function
warm=np.maximum(0,np.minimum(1,np.minimum((x-10)/(20-10),(30-x)/(30-25))))

#Define fuzzy set "hot" using a trpezoidal membership function
hot=np.maximum(0,np.minimum(1,np.maximum((x-25)/(30-25),0)))

#plot fuzzy sets "cold","warm",and "hot"
plt.figure(figsize=(12,6))
plt.plot(x,cold,label='cold',linewidth=2)
plt.plot(x,warm,label='warm',linewidth=2)
plt.plot(x,hot,label='hot',linewidth=2)
plt.xlabel('Temperature (c)')
plt.ylabel('Membership Degree')
plt.title('Fuzzy sets for temperature levels')
plt.legend()
plt.grid(True)
plt.show()
```
