```python
import numpy as np
import matplotlib.pyplot as plt

#define fuzzy sets using membership functions
x = np.arange(0,10,0.1)
A = np.maximum(0,np.minimum(1,(x-2)/(6-2)))
B = np.maximum(0,np.minimum(1, (8-x)/(8-4)))
```

```python
# plot fuzzy sets A and B
plt.figure(figsize=(12,6))
plt.plot(x,A,label='set A')
plt.plot(x,B,label='set B')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title('Fuzzy sets A and B')
plt.legend()
plt.show()
```

```python
# union of fuzzy sets A and B
union = np.maximum(A,B)
```

```python
#plot the result
plt.figure(figsize=(12,6))
plt.plot(x,A, label='set A')
plt.plot(x,B,label='set B')
plt.plot(x,union,label='Union A U B',linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title("Union of fuzzy sets A and B")
plt.legend()
plt.show()
```

```python
# Intersection of fuzzy sets A and B
intersection = np.minimum(A, B)

# Plot the result
plt.figure(figsize=(12, 6))
plt.plot(x, A, label='Set A')
plt.plot(x, B, label='Set B')
plt.plot(x, intersection, label='Intersection A ∩ B', linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title('Intersection of Fuzzy Sets A and B')
plt.legend()
plt.show()
```

```python
#compliment of fuzzy set A
compliment_A=1-A

#plot the result
plt.figure(figsize=(12,6))
plt.plot(x,A,label='set A')
plt.plot(x,compliment_A,label='compliment of A',linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title("compliment of fuzzy set A")
plt.legend()
plt.show()
```

```python
# general aggregation operation (e.g., Algebraic)
aggregation=A+B-A*B

#plot the result
plt.figure(figsize=(12,6))
plt.plot(x,A,label='set A')
plt.plot(x,B,label='set B')
plt.plot(x,aggregation,label='Aggregation (Algebraic sum)',linestyle='dashed')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Degree')
plt.title("Aggregation operation on fuzzy sets A and B ")
plt.legend()
plt.show()
```
