```python
import numpy as np
import matplotlib.pyplot as plt

#Define the range for the universe of discourse
x=np.arange(0,100,0.1)

#Define fuzzy set "cold" using a trpezoidal membership function
Slow=np.maximum(0,np.minimum(1,np.minimum((30-x)/(20-10),1)))

#Define fuzzy set "Warm" using a trpezoidal membership function
Medium=np.maximum(0,np.minimum(1,np.minimum((x-30)/(20-10),(60-x)/(60-50))))

#Define fuzzy set "hot" using a trpezoidal membership function
Fast=np.maximum(0,np.minimum(1,np.maximum((x-60)/(90-80),0)))

#plot fuzzy sets "Slow","Medium",and "Fast"
plt.figure(figsize=(40,10))
plt.plot(x,Slow,label='Slow',linewidth=2)
plt.plot(x,Medium,label='Medium',linewidth=2)
plt.plot(x,Fast,label='Fast',linewidth=2)
plt.xlabel('Speed (Kmph)')
plt.ylabel('Membership Degree')
plt.title('Fuzzy sets for Speed levels')
plt.legend()
plt.grid(True)
plt.show()
```
