```python
#Fuzzy Logic Classification of Height Using Trapezoidal and Triangular Membership Functions

import numpy as np
import matplotlib.pyplot as plt

# Define the range for the universe of discourse (Height in cm)
x = np.arange(100, 200, 0.1)

# Define a trapezoidal membership function manually
def trapezoidal_mf(x, a, b, c, d):
    return np.maximum(0, np.minimum((x - a) / (b - a), np.minimum(1, (d - x) / (d - c))))

# Define fuzzy set "Short" using a manually defined trapezoidal membership function
short = trapezoidal_mf(x, 100, 120, 140, 160)

# Define fuzzy set "Average" using a manually defined triangular membership function
def triangular_mf(x, a, b, c):
    return np.maximum(0, np.minimum((x - a) / (b - a), (c - x) / (c - b)))

average = triangular_mf(x, 140, 160, 180)

# Define fuzzy set "Tall" using a manually defined trapezoidal membership function
# Adjust the parameters to avoid division by zero
tall = trapezoidal_mf(x, 160, 180, 190, 200)

# Plot fuzzy sets "Short", "Average", and "Tall"
plt.figure(figsize=(12, 6))
plt.plot(x, short, label='Short', linewidth=2)
plt.plot(x, average, label='Average', linewidth=2)
plt.plot(x, tall, label='Tall', linewidth=2)
plt.xlabel('Height (cm)')
plt.ylabel('Membership Degree')
plt.title('Fuzzy Sets for Height Levels')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate a specific height
test_height = 165  # Example height value
membership_short = trapezoidal_mf(test_height, 100, 120, 140, 160)
membership_average = triangular_mf(test_height, 140, 160, 180)
membership_tall = trapezoidal_mf(test_height, 160, 180, 190, 200)

print(f'For a height of {test_height} cm:')
print(f'Membership degree in Short: {membership_short:.2f}')
print(f'Membership degree in Average: {membership_average:.2f}')
print(f'Membership degree in Tall: {membership_tall:.2f}')

# Visualize the membership degrees for the specific height
plt.figure(figsize=(12, 6))
plt.plot(x, short, label='Short', linewidth=2)
plt.plot(x, average, label='Average', linewidth=2)
plt.plot(x, tall, label='Tall', linewidth=2)
plt.axvline(x=test_height, color='r', linestyle='--', label=f'Test Height: {test_height} cm')
plt.xlabel('Height (cm)')
plt.ylabel('Membership Degree')
plt.title('Membership Degrees for Test Height')
plt.legend()
plt.grid(True)
plt.show()
```
